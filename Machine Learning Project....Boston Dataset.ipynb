# Train the model

```python
x = Boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis',
       'rad', 'tax', 'ptratio', 'black', 'lstat']]
y = Boston['medv']
```

# Test the model
